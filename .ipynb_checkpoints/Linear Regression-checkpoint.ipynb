```python
import numpy as np
from numpy import arange
import pandas as pd
import random
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from pylab import plot, show, xlabel, ylabel
```

```python
#Load Dataset
#boston = load_boston()
X = load_boston()
y = X['target']
#pd.DataFrame(X.data)
# #Add bias term in X
x=X.data
X = np.append(np.ones((X.data.shape[0],1)),X.data,1)
# #make y as nd-arry

y = np.reshape(y,(-1,1))
m=X.shape[0]
n =X.shape[1]
print(m)
print(n)
#pd.DataFrame(X)
#pd.DataFrame(y)
#print(y)
pd.DataFrame(X)
```

```python
#divide data into train,validate and test as 60%,20%,20%

train_start = 0
train_end = int(.6*m)
validation_start = int(.6*m)
validation_end = int(.8*m)
test_start = int(.8*m)
test_end = m

X_train = X[train_start:train_end,:]
y_train = y[train_start:train_end,:]

X_validate = X[validation_start:validation_end,:]
y_validate = y[validation_start:validation_end,:]

X_test = X[test_start:test_end,:]
y_test = y[test_start:test_end,:]

print (X.shape,y.shape)
print (X_train.shape,y_train.shape)
print (X_validate.shape,y_validate.shape)
print (X_test.shape,y_test.shape)

assert X_train.shape[0]+X_validate.shape[0]+X_test.shape[0] == X.shape[0]
assert y_train.shape[0]+y_validate.shape[0]+y_test.shape[0] == y.shape[0]
```

```python
#X dim: m+1*(n+1) where n = no. of fetaures
#y dim: m*1
#theta dim: (n+1)*1
def gradientDescent(X,y,theta,m,alpha,numIter):
    X_transpose = X.transpose()
    prevcost = 1e8
    c=[]
    for i in range(numIter):
        c.append(0)
    for i in range(numIter):
        hypotheis = np.dot(X,theta)
        loss = hypotheis - y
        cost = np.sum(loss**2)/(2*m)
        if cost > prevcost:
            "error!! not converging"
            break
        gradient = np.dot(X_transpose,loss)/m
        theta = theta - alpha*gradient
        prevcost = cost
        c[i]=cost
        if i == numIter-1:
            print("Loss: ",cost)
    plot(arange(numIter),c)
    xlabel('Iterations')
    ylabel('Cost Function')
    show()
    return theta
```

```python
#initialize theta
theta = np.ones((n,1))
theta = gradientDescent(X_train,y_train,theta,X_train.shape[0],.01e-4,100)
#theta = gradientDescent(X_train,y_train,theta,X_train.shape[0],.01e-4,100000)
print("theta values")
print(theta)
```

```python

def predict(X,theta):
    return np.dot(X,theta)
```

```python
#R^2 Score
def score(X,y,theta):
    y_pred = predict(X,theta)
    y_true = y
    u = np.sum((y_pred-y_true)**2)
    v = np.sum((y_true - np.mean(y_true))**2)
    return 1-(u/v)
```

```python
print(score(X_test,y_test,theta))
```
